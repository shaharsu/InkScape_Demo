````python
import matplotlib.pyplot as plt
import numpy as np
````

## Generate data for plotting

````python
x=np.linspace(0, 10, 100)
y = np.random.randn(100)+0.5
````

## Plot data with default values
Note that by default each letter is rendered individually and text is not editable. An svg file is saved by matplotlib, to be imported into inkscape

````python
fig,ax = plt.subplots(figsize=[5,5])
ax.scatter(x, y, color='blue', edgecolor='black', s=50, alpha=0.5)
ax.set_xlabel('X-axis label')
ax.set_ylabel('Y-axis label')
ax.set_title('Scatter Plot Example')
fig.savefig('scatter_plot_exampe.svg')
````

## Set default values for matplotlib
We use the plt.rcParams to set default values for the plot. This will affect all plots generated after this point.
Most important is the 'svg.fonttype' which is set to 'none'. This will ensure that text is rendered as text in the svg file, and not as paths. This allows for editing the text in inkscape.
```

````python
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams['axes.linewidth'] = 3 #set the value globally
plt.rcParams['xtick.major.size'] = 10
plt.rcParams['xtick.major.width'] = 3
plt.rcParams['xtick.minor.size'] = 5
plt.rcParams['xtick.minor.width'] = 3
plt.rcParams['ytick.major.size'] = 10
plt.rcParams['ytick.major.width'] = 3
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.minor.size'] = 5
plt.rcParams['ytick.minor.width'] = 3
plt.rcParams['font.size'] = 20
````

## replotting the figure with new defaults
File will be saved as a new svg to be imported into inkscape.

````python
fig,ax = plt.subplots(figsize=[5,5])
ax.scatter(x, y, color='blue', edgecolor='black', s=50, alpha=0.5)
ax.set_xlabel('X-axis label')
ax.set_ylabel('Y-axis label')
ax.set_title('Scatter Plot Example')
fig.savefig('scatter_plot_example_2.svg')
````
